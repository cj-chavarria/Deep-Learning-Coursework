# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# Parte 1

## 1. Abrir una imagen

### 1.1 "Hola mundo“ (Método oficial)

```python
# Al ejecutar esta celda se abre una ventana externa, es necesario cerrar esa ventana
# para continuar la ejecución de las siguientes celdas.
img =  cv.imread('figuras/paisaje.jpg')

cv.imshow('Imagen Original', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 1.2 "Hola mundo“ (Recomendado para openCV en Jupyter)

```python
img =  cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img.shape
```

## 2. Conversión a escala de grises en openCV 

```python
img_gray =  cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gray[50:100,100:200] = 0
img_gray[100:150,100:200] = 255
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
img_gray.shape
```

```python
img_arco = img_gray[100:350, 200:400]

plt.imshow(img_arco, cmap='gray')
plt.show()
```

## 3.Modificando partes de una imagen

## 4. Filtros

```python
img_fmedia = cv.blur(img_gray, (7,7))

plt.imshow(img_fmedia, cmap='gray')
plt.show()
```

```python
img_fgauss = cv.GaussianBlur(img_gray, (7,7), 0)

plt.imshow(img_fgauss, cmap='gray')

plt.show()
```

### 4.1 Ejemplo de aplicación del filtro Mediana

```python
img_fmediana = cv.medianBlur(img_gray, 9)

plt.imshow(img_fmediana, cmap='gray')
plt.show()
```

```python
cerebro =  cv.imread('figuras/cerebro.png')
cerebro_gray = cv.cvtColor(cerebro, cv.COLOR_BGR2GRAY)
plt.imshow(cerebro, cmap='gray')
plt.show()
```

```python
cerebro_fmediana = cv.medianBlur(cerebro_gray, 7)

plt.imshow(cerebro_fmediana, cmap='gray')
plt.show()
```

## 5. Thresholding 

```python
th, img_th = cv.threshold(img_gray, 150, 255, cv.THRESH_BINARY)
plt.imshow(img_th, cmap='gray')
plt.show()
```

## 6. Resize o cambio de tamaño

```python
fracc = 0.25

img_resize = cv.resize(img_th, ( int(img_th.shape[1]*fracc), int(img_th.shape[0]*fracc) ) )
plt.imshow(img_resize, cmap='gray')
plt.show()
```
