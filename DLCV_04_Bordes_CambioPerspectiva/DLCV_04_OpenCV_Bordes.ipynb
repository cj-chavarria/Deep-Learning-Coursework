# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
# Cargar la imagen 'figuras/coche.jpg' en escala de grises
img = cv.imread('figuras/coche.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Aplicar filtro Gaussiano para suavizar la imagen
img_gauss = cv.GaussianBlur(img_gray, (7,7), 0)

plt.imshow(img_gauss, cmap='gray')
plt.show()
```

```python
# Aplicar el detector de bordes de Canny a la imagen en escala de grises
img_border = cv.Canny(img_gray, 100,200)

plt.imshow(img_border, cmap='gray')
plt.show()
```

```python
# Filtro Gaussiano + Bordes con Canny
img_border_gauss = cv.Canny(img_gauss, 100,200)

plt.imshow(img_border_gauss, cmap='gray')
plt.show()
```

## Detección de bordes con Canny

```python

# Mostrar las imágenes
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.title('Imagen Original en escala de grises')
plt.imshow(img_gray, cmap='gray')
 
plt.subplot(2,2,2)
plt.title('Filtro Gaussiano')
plt.imshow(img_gauss, cmap='gray')
 

plt.subplot(2,2,3)
plt.title('Bordes con Canny ')
plt.imshow(img_border, cmap='gray')
 

plt.subplot(2,2,4)
plt.title('Filtro Gaussiano + Bordes con Canny')
plt.imshow(img_border_gauss, cmap='gray')
plt.show()
```
