```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# 1. Operaciones lógicas


 ## Implementación de operadores OpenCV AND, OR, XOR y NOT  

<div class="alert alert-info">
 <ul>
  <li>Generar imagenes como las que se muestran a continuación (300 x 300)</li>
</ul> 

</div>
<img align="left" width="500"  float= "none" align="middle" src="figuras/rectangulo_circulo.png">

```python
# Rectangulo 

rectangulo =  np.zeros((300,300), dtype='uint8')
cv.rectangle(rectangulo, (25,25), (275,275), 255, -1)

plt.imshow(rectangulo, cmap='gray')
plt.show()
```

```python
# Circulo
circulo =  np.zeros((300,300), dtype='uint8')
cv.circle(circulo, center=(150,150), radius=150, color=255, thickness=-1)
plt.imshow(circulo, cmap='gray')
plt.show()
```

<div class="alert alert-info">
 <ul>
  <li>Aplicar operaciones de AND, OR y XOR y NOT</li>
</ul> 


```python
#AND
img_and = cv.bitwise_and(rectangulo, circulo)
plt.imshow(img_and, cmap='gray')
plt.show()
```

```python
#OR
img_or =  cv.bitwise_or(rectangulo, circulo)
plt.imshow(img_or, cmap='gray')
plt.show()
```

```python
#XOR
img_xor = cv.bitwise_xor(rectangulo, circulo)
plt.imshow(img_xor, cmap='gray')
plt.show()
```

<div class="alert alert-info">
 <ul>
  <li>Aplicar la operacón inversa a ambas imagenes</li>
</ul> 

```python
#Rectangulo
rectangulo_inv = cv.bitwise_not(rectangulo)
plt.imshow(rectangulo_inv, cmap='gray')
plt.show()
```

```python
#Circulo
circulo_inv =  cv.bitwise_not(circulo)
plt.imshow(circulo_inv, cmap='gray')
plt.show()
```

# 2. Máscaras

```python
# Abrir el archivo 'figuras/people.jpg'

img =  cv.imread('figuras/people.jpg')
img =  cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
# Mascara rectangular
mascara_rectangular =  np.zeros((400, 800), dtype='uint8')
cv.rectangle(mascara_rectangular, (100, 50), (220, 250), 255, -1)
plt.imshow(mascara_rectangular, cmap='gray')
plt.show()
```

```python
#Aplicar la mascara rectangular

img_mascara_rectangular = cv.bitwise_and(img, img, mask=mascara_rectangular)
plt.imshow(img_mascara_rectangular)
plt.show()
```

```python
# Mascara circular y su implementación
mascara_circular =  np.zeros((400, 800), dtype='uint8')
cv.circle(mascara_circular, center=(150,150), color=255, radius=100, thickness=-1)
plt.imshow(mascara_circular, cmap='gray')
plt.show()
```

```python
img_mascara_circular = cv.bitwise_and(img, img, mask=mascara_circular)
plt.imshow(img_mascara_circular)
plt.show()
```

# Opening

```python
#Opening
img_gris = cv.imread('figuras/labdatos.png')
img_gris = cv.cvtColor(img_gris, cv.COLOR_BGR2GRAY)

kernel =  cv.getStructuringElement(cv.MORPH_RECT, (5,5))
img_opening = cv.morphologyEx(img_gris, cv.MORPH_OPEN, kernel)

plt.imshow(img_opening, cmap='gray')
plt.show()
```

## Closing

```python
#Closing
img_gris = cv.imread('figuras/labdatos.png')
img_gris = cv.cvtColor(img_gris, cv.COLOR_BGR2GRAY)
kernel =  cv.getStructuringElement(cv.MORPH_RECT, (6,6))
img_closing = cv.morphologyEx(img_gris, cv.MORPH_CLOSE, kernel)
plt.imshow(img_closing, cmap='gray')
plt.show()
```

# Máscara para un rango de color

```python
#Mostramos la imagen original

img = cv.imread('figuras/flor_morada.jpg')
img_rgb =  cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

```python
#Llevamos a cabo el cambio de espacio de color y definimos los rangos 
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
inferior = np.array([109, 30,0])
superior =  np.array([179,255,255])
```

```python
plt.imshow(img_hsv)
```

```python
#Aplicamos la mascara de la region de color

mascara =  cv.inRange(img_hsv, inferior, superior)
```

```python
# Revisamos los valores de los pixeles
mascara.shape
```

```python
# Visualizamos la mascara
plt.imshow(mascara, cmap='gray')
plt.show()
```

```python
#Obtenemos la imagen en escala de grisis y eliminamos los valores de la mascara
img_gris=  cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gris_sub =  cv.subtract(img_gris, mascara)
plt.imshow(img_gris_sub, cmap='gray')
plt.show()
```

```python
#COnvertimos la imagen resultante en RGB

img_gris_sub_rgb = cv.cvtColor(img_gris_sub, cv.COLOR_GRAY2BGR)
plt.imshow(img_gris_sub_rgb)
img_gris_sub_rgb.shape
```

```python
#Obtenemos la mascara aplicada a la imagen real
img_mascara = cv.bitwise_and(img_rgb, img_rgb, mask=mascara)
plt.imshow(img_mascara)
```

```python
#Resaltando un color en una imagen  
img_final = cv.add(img_mascara, img_gris_sub_rgb)
plt.imshow(img_final)
plt.show()
```
