# 1. Figuras geométricas


### Creación de una imagen en color negro

```python
# Imagen color negro de 512 x 512 
img_base = np.zeros(shape=(512,512,3), dtype='uint8')

plt.imshow(img_base)
plt.show()
```

#### Rectangulo

```python
#cv.rectangle(imagen,origen,final,color,grosor de linea)
img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.rectangle(img_base, (0,0), (100,100), (0,0,255), thickness=5)
cv.rectangle(img_base, (100,300), (200,400), (255,0,255), thickness=5)

plt.imshow(img_base)
plt.show()
```

#### Circulo

```python
#cv.circle(imagen, centro, radio, color, grosor de linea)
img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.circle(img_base, center=(512//2,512//2), radius=512//2, color=(255,0,0), thickness=5)
plt.imshow(img_base)
plt.show()
```

#### Linea

```python
#cv.line(imagen, inicio, fin, color, grosor de linea)
img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.line(img_base, pt1=(0,0), pt2=(512,512), color=(0,255,0), thickness=5)

plt.imshow(img_base)
plt.show()
```

```python
img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.line(img_base, pt1=(0,512//2), pt2=(512,512//2), color=(0,0,255), thickness=5)

plt.imshow(img_base)
plt.show()
```

#### Texto

```python
font = cv.FONT_HERSHEY_SIMPLEX
lineType=cv.LINE_AA
#cv.putText(imagen,texto,origen, fuente,tamaño de la fuente,color,grosor, tipo de linea)

img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.putText(img_base, text='hola mundo', org=(128,256), fontFace=font, fontScale=1.5, color=(0,255,0), thickness=2, lineType=lineType)

plt.imshow(img_base)
plt.show()
```


#### Polígono 

```python
#cv.polylines(imagen,puntos,Cerrada o abierta,color y grosor)
puntos=np.array( [ [100,200],[400,200],[400,500],[200,400]  ]  )

img_base = np.zeros(shape=(512,512,3), dtype='uint8')

cv.polylines(img_base, pts=[puntos], isClosed=True, color=(255,0,0), thickness=5)

plt.imshow(img_base)
plt.show()
```

## 4. Detección de contornos

```python
#Visualizar figuras/coins.png
img_coins =  cv.imread('figuras/coins.png')
img_coins = cv.cvtColor(img_coins, cv.COLOR_BGR2RGB)
plt.imshow(img_coins)
```

```python
#Pasar a escala de grises
img_gray = cv.cvtColor(img_coins, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
#Aplicar el filtro GaussianBlur
img_fgauss =  cv.GaussianBlur(img_gray, (3,3), 0)
plt.imshow(img_fgauss, cmap='gray')
```

```python
#Aplicar la binarización con la opción cv.THRESH_BINARY_INV
thr, img_bin = cv.threshold(img_fgauss, 155, 255, cv.THRESH_BINARY_INV)
plt.imshow(img_bin, cmap='gray')
```

```python
#Encontrar los contornos con cv.findContours(IMAGEN-BN, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
contornos, h = cv.findContours(img_bin, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
print(len(contornos))
```

```python
# Visualizar los contornos  con drawContours
cv.drawContours(img_coins, contornos, -1, (0,255,0), 2)
plt.imshow(img_coins)
```
