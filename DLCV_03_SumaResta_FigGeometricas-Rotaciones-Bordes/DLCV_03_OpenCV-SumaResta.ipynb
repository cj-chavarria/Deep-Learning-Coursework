# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# Parte II

### 1.Operaciones aritméticas

Las imágenes son arreglos de NumPy almacenados como enteros de 8 bits sin signo (unidad8) con valores en el rango [0, 255]; al usar las funciones de add/subtract en OpenCV, estos valores se *acoplarán* a este rango, incluso si caen fuera del rango [0, 255] después de aplicar la operación

```python
a = np.array([[200]], dtype=np.uint8)
b = np.array([[100]], dtype=np.uint8)

print("a:", a, a.dtype, a.shape)
print("b:", b, b.dtype, b.shape)

added = cv.add(a, b)
subtracted = cv.subtract(np.array([[50]], dtype=np.uint8), b)

print("added:", added, added.dtype, added.shape)
print("subtracted:", subtracted, subtracted.dtype, subtracted.shape)
```

```python
#Cargar la imagen figuras/lago.jpg
img = cv.imread('figuras/lago.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

### 2. Aumentar las intensidad

```python
# Aumentar las intensidades de píxeles en nuestra imagen en 100
M = np.ones(img.shape, dtype = np.uint8) * 100

img_suma = cv.add(img_rgb, M)

plt.imshow(img_suma)
plt.show()
```

### 3. Disminuir la intensidad

```python
# Disminuir la intensidade de píxeles en nuestra imagen con 50
M = np.ones(img.shape, dtype = np.uint8) * 50
img_resta = cv.subtract(img_rgb, M)
plt.imshow(img_resta)
plt.show()
```

### 4. Alta y baja intensidad

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,20))

ax1.imshow(img_rgb)
ax2.imshow(img_suma)
ax3.imshow(img_resta)

plt.show()
```
